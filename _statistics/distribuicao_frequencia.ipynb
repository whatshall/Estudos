```python
## DISTRIBUIÇÃO DE FREQUÊNCIAS
import numpy as np
import pandas as pd
from collections import Counter
```

```python
dados_sp = pd.read_csv('C:/Users/leand/Documents/scripts_jupyter/bases/enem_2019_tratado.csv',
                    sep=',', 
                    encoding='iso-8859-1')
```

```python
enem_sjrp = dados_sp.loc[dados_sp.NO_MUNICIPIO_RESIDENCIA == 'São José do Rio Preto']
```

```python
enem_sjrp.head()
```

```python
enem_sjrp.shape
```

```python
idades = enem_sjrp['IDADE']
```

```python
idades
```

```python
freq_absoluta = Counter(idades)
```

```python
freq_absoluta
```

```python
freq_absoluta = pd.DataFrame.from_dict(freq_absoluta, orient='index')
```

```python
freq_absoluta
```

```python
freq_absoluta = freq_absoluta.sort_index(ascending=True)
freq_absoluta
```

```python
freq_relativa = freq_absoluta / freq_absoluta.sum()
freq_relativa
```

```python
freq_relativa_perc = freq_relativa * 100
freq_relativa_perc
```
